```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
from collections import Counter
```

```python
complaints = pd.read_csv("consumer_complaints.csv", dtype=str)
```

```python
complaints.head()
```

```python
complaints['Date received'] = pd.to_datetime(complaints['Date received'])
complaints['Date sent to company'] = pd.to_datetime(complaints['Date sent to company'])
```

```python
#complaints.index = complaints.pop('Complaint ID')
```

```python
complaints.head()
```

```python
date_group = complaints.groupby('Date received').agg(['count'])
date_group = date_group.pop('Complaint ID')
```

```python
date_group.head()
```

```python
date_group = date_group.sort(['count'], ascending=[False]).sort(['count'], ascending=[False])
date_group.head(10)
#This was cooler with the mean for the days in the month, but can't sort muti-index
```

```python
date_group.resample('M', how=["sum"])[:10]
```

```python
products_group = complaints.groupby('Product').agg(['count'])
products_group = products_group.pop('Complaint ID')
products_group.sort(['count'], ascending=[False]).sort(['count'], ascending=[False])
```

```python
company_group = complaints.groupby('Company').agg(['count'])
company_group = company_group.pop('Complaint ID')
company_group = company_group.sort(['count'], ascending=[False]).sort(['count'], ascending=[False])
company_group.head(10)
```

```python
response_group = complaints.groupby('Company response').agg(['count'])
response_group = response_group.pop('Complaint ID')
response_group = response_group.sort(['count'], ascending=[False]).sort(['count'], ascending=[False])
response_group
```
