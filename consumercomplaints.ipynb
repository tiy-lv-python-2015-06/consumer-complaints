```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_complaints = pd.read_csv("/users/scottholmes/Documents/complaints/consumer_complaints.csv", header=0, dtype=str)
```

```python
df_complaints.info()
```

```python
df_complaints["temp_date_received"] = df_complaints['Date received']
df_complaints["temp_product"] = df_complaints['Product']
df_complaints["temp_company"] = df_complaints['Company']
df_complaints["temp_company_response"] = df_complaints['Company response']
```

```python
df_complaints.head()
```

```python
df_complaints
```

```python
# Below is the count of the product complaints along with its graph.  Mortgages seemed to have the most complaints
# Prepaid cards having the least complaints.
```

```python
product_complaints = df_complaints.groupby('Product')
count_product_complaints = product_complaints.count()
```

```python
print(count_product_complaints["Complaint ID"])
```

```python
count_product_complaints["Complaint ID"].plot()
```
