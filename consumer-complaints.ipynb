```python
maindf = pd.read_csv("Consumer_Complaints.csv", dtype=str)
```

```python
maindf['Complaint ID'] = maindf['Complaint ID'].astype(np.int64)
```

```python
date_column = maindf.pop('Date received')
```

```python
maindf['Date received'] = pd.to_datetime(date_column, format="%m/%d/%Y")
```

```python
comp_by_month = maindf
```

```python
comp_by_month.index = maindf['Date received']
```

```python
comp_by_month_df = maindf
```

```python
comp_by_month_df = comp_by_month_df.groupby(lambda x: x.month).count()
```

```python
comp_by_month_df
```

```python
comp_by_product_df = maindf.groupby("Product").count()
```

```python
comp_by_product_df
```

```python
comp_by_company_df = maindf.groupby("Company").count()
```

```python
comp_by_company_df.sort('Complaint ID')[:-10:-1]
```

```python
comp_by_company_df[:-10:-1]
```

```python
comp_by_company_res_df = maindf.groupby("Company response", sort=True).count()
```

```python
comp_by_company_res_df
```

```python
comp_by_DOW_df = maindf
comp_by_DOW_df = comp_by_DOW_df.resample('weekday', how="count")
```

```python
comp_by_DOW_df = maindf
comp_by_DOW_df = comp_by_DOW_df.groupby(lambda x: x.weekday).count()
```
